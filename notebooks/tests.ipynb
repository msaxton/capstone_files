#### Checking corpus to see how many english and german articles there are

```python
eng_count = 0
ger_count = 0
other_count = 0
pub_years = []
total = len(os.listdir(path))
for file in os.listdir(path):
    tree = ET.parse(path + file)
    root = tree.getroot()
    year = root.find('./front/article-meta/pub-date/year')
    pub_years.append(year.text)
    lang = root.find('./front/article-meta/custom-meta-group/custom-meta/meta-value')
    if (lang.text == 'eng') or (lang.text == 'en'):
        eng_count += 1
    elif lang.text == 'ger':
        ger_count += 1
    else:
        other_count += 1
print('total -> ', total)
print('eng -> ', eng_count)
print('ger -> ', ger_count)
print('other -> ', other_count)
```
